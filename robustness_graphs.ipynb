```python
name= "cub" 

# Carica i dati
df = pd.read_csv(StringIO(data_cub))
df['geometry'] = df['exp'].str.split('_').str[0]



# Calcola media e std
agg_df = df.groupby('geometry')[['eps0', 'eps1', 'eps2', 'eps3']].agg(['mean', 'std'])

# Costruisci nuovo dataframe con valori formattati
result_df = pd.DataFrame(index=agg_df.index)

for col in ['eps0', 'eps1', 'eps2', 'eps3']:
    means = agg_df[(col, 'mean')]
    stds = agg_df[(col, 'std')]
    result_df[col] = [f"{m:.4f} ± {s:.4f}" for m, s in zip(means, stds)]

# Salva su CSV
result_df.to_csv(f"pgd_eps_{name}.csv")

# Mostra il risultato
print(result_df)
```
